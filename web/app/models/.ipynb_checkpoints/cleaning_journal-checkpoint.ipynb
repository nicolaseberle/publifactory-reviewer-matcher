```python
import pandas as pd
```

```python
df = pd.read_csv("w2v/scimagojr 2018.csv")
```

```python
print(df.columns)
```

```python
print(df.head())
```

```python
df1 = df[['Title', 'Issn', 'Total Docs. (2018)', 'Categories']]
```

```python
df1 = df1.rename(columns={"Total Docs. (2018)": "h-index"})
```

```python
df1['Categories'] = df1['Categories'].str.split(';')
```

```python
print(df1)
```

```python
new_cat = []

for line in df1.Categories:
    temp = []
    if type(line) == type([]):
        for val in line:
            val = val.replace(' (Q1)', '')
            val = val.replace(' (Q2)', '')
            val = val.replace(' (Q3)', '')
            val = val.replace(' (Q4)', '')
            val = val.replace(' (miscellaneous)', '')
            if val.startswith(' '):
                val = val[1:]
            temp.append(val)
    new_cat.append(temp)
```

```python
df1['Categories'] = new_cat
```

```python
print(df1.Categories)
```

```python
result = {}
for line in df1.Categories:
    if type(line) == type([]):
        for val in line:
            if val not in result:
                result[val] = 1
            else:
                result[val] += 1

print(result)
```

```python
sorted_result = sorted(result.items(), key=lambda kv: kv[1], reverse=True)

print(sorted_result)
```

```python
print(len(sorted_result))
```
